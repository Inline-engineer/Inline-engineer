```python
import pandas as pd

df = pd.read_csv('/content/archive_extracted/Unemployment in India.csv')

# Clean columns
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Convert date with explicit format to avoid warning and handle whitespace
df['date'] = pd.to_datetime(df['_date'].str.strip(), format='%d-%m-%Y')

# Handle missing values
df = df.dropna()

df.head()
```

```python
df.describe()
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df['date'], df['_estimated_unemployment_rate_(%)'])
plt.title("Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Rate (%)")
plt.show()
```

```python
covid = df[(df['date'] >= '2020-01-01') & (df['date'] <= '2021-12-31')]
pre = df[df['date'] < '2020-01-01']

print("Pre-COVID:", pre['_estimated_unemployment_rate_(%)'].mean())
print("COVID:", covid['_estimated_unemployment_rate_(%)'].mean())
```

```python
df['month'] = df['date'].dt.month
monthly_avg = df.groupby('month')['_estimated_unemployment_rate_(%)'].mean()

monthly_avg.plot(kind='bar', title='Monthly Unemployment Pattern')
plt.show()
```

```python
# Save cleaned data
df.to_csv('cleaned_data.csv', index=False)

# Save figure
plt.figure(figsize=(10,5))
plt.plot(df['date'], df['_estimated_unemployment_rate_(%)'])
plt.savefig('trend.png')
```
